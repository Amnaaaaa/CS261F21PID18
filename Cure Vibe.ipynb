```python
import requests
from bs4 import BeautifulSoup
import pandas as pd



Name = []
Quantity = []
price = []
discription = []
category = []
discount = []
ingredients = []

url = "https://www.1mg.com/"
data = requests.get(url)
dataContent = data.content


urls = ["https://www.1mg.com/categories/homeopathy-57" , "https://www.1mg.com/categories/featured/monsoon-store-854","https://www.1mg.com/categories/ayurveda/general-health-267"]
categories = ["Homeopathy", "moonsoon","Ayurveda"]
endOfPages = [850,525,472]


    

startCategory = 0
srartingPage = 1



# startCategory = url.readline()



# f = open("demofile3.txt", "w")
# f.write("Woops! I have deleted the content!")
# f.close()

# #open and read the file after the appending:
# f = open("demofile3.txt", "r")
# print(f.read())

source = requests.get(urls[startCategory])
source = source.content
source = BeautifulSoup(source, 'html.parser')

urlr = open("url.txt","r")
startCategory = int(urlr.read())

spage = open("spage.txt","r")
srartingPage = int(spage.read())

for index in range(startCategory, len(urls)):
    
    
    print("--------------Starting with category "+categories[index]+"--------------")

    for i in range(srartingPage, endOfPages[index]+1):
        
        print("category = "+ str(categories[index]) + "Page No. "+ str(i)+" out of " +str(endOfPages[index]))

        divs = source.find_all('div' , class_ = 'col-md-3 col-sm-4 col-xs-6 style__container___1TL2R')

        for div in divs:
            name = div.find('div', class_= 'style__pro-title___2QwJy').text
            Name.append(name)
            q = div.find('div', class_ = 'style__pack-size___2JQG7').text
            Quantity.append(q)
            p = div.find('div', class_ = 'style__price-tag___cOxYc').text
            price.append(p)
            discription.append(str(name)+"\n"+str(q)+"\n")

            category.append(categories[index])

            try:
                d = div.find('span',class_ = 'style__off-badge___2JaF-').text
            except:
                d = ""

            discount.append(d)
            ingredients.append(q)

        entities = {
            'Name': Name,
            'Quantity': Quantity,
            'Price': price,
            'Discription': discription,
            'Category': category,
            'Discount': discount,
            'Ingredients':ingredients 
        }
        df = pd.DataFrame(entities)

        df.to_csv('medicineScrapping.csv', mode = 'a', index = False,header = False)
        
        r = open ("spage.txt","w")
        r.write(str(i+1))
        r.close()
        
        
        Name = []
        Quantity = []
        price = []
        discription = []
        category = []
        discount = []
        ingredients = []

        url2 = url+"?filter=true&pageNumber="+ str(i)
        source = requests.get(url2)
        source = source.content
        source = BeautifulSoup(source, 'html.parser')
        
        
    u = open ("url.txt","w")

    u.write(str(index+1))
    u.close()
    srartingPage = 1
    
    source = requests.get(urls[index])
    source = source.content
    source = BeautifulSoup(source, 'html.parser')




```
